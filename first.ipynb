```python
import pandas as pd
filepath = "./novel-corona-virus-2019-dataset/time_series_covid19_confirmed_global.csv"

def get_novel(fp):
    data = pd.read_csv(fp)

    # clean data 
    data = data.drop(['Lat', 'Long'], axis=1)
    data = data.groupby(['Country/Region']).sum()
    data = data.reset_index()
    return data

data = get_novel(filepath)

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
data
```

```python
N_COUNTRIES = 20

def get_top_countries(data, n_c):
    last_h = data.columns[-1]
    data = data.sort_values(by=last_h, ascending=False)
    data = data.head(n_c)
    return data


# sort by total number of confirmed (descending)
data = get_top_countries(data, N_COUNTRIES)
data
```

```python
import os
import csv
TRESHOLD = 100

# empty new dataset
header = [i for i in range(data.shape[1]-1)]
header.insert(0, "country")
new_data = pd.DataFrame(columns=header)

# populate new dataset -> it contains the cumulative daily number of confirmed after the reaching of 100 cases.
for _, row in data.iterrows():
    country = row[0]
    prog_confirmed = row[1:]
    starting_date = len(prog_confirmed)
    for i, p in enumerate(prog_confirmed):
        if p >= TRESHOLD:
            starting_date = i
            break
    if starting_date < len(prog_confirmed):
        new_row = [r for r in prog_confirmed[starting_date:]]
        new_row.insert(0, country)
        for j in range(starting_date):
            new_row.append(-1)
        new_row = pd.DataFrame([new_row], columns=header)
        new_data = new_data.append(new_row, ignore_index=True)

def write_dataset(dataset, filepath):
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    data_file = open(filepath, mode='w')
    file_writer = csv.writer(data_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    file_writer.writerow(dataset.columns)
    for _, d in dataset.iterrows():
        file_writer.writerow(d)
    data_file.close()
    
filepath_new_data_csv = "./novel-corona-virus-2019-dataset/new_data.csv"
write_dataset(new_data, filepath_new_data_csv)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def to_plot(x, y):
    return [(xx, yy) for xx, yy in zip(x, y) if yy>=0]

def plot(dataset):
    i_mark = 0
    for index_country, row_country in dataset.iterrows():
        data_to_plot = to_plot(row_country.index, row_country)
        sns.lineplot(x=[t[0] for t in data_to_plot], y=[t[1] for t in data_to_plot], label=index_country, 
                     sort=False, marker='o' if i_mark>=10 else '')
        i_mark = i_mark+1
        
def plot_smooth(dataset, degree):
    i_mark = 0
    for index_country, row_country in dataset.iterrows():
        data_to_plot = to_plot(row_country.index, row_country)
        list_x = [int(t[0]) for t in data_to_plot]
        list_y = [t[1] for t in data_to_plot]
        poly = np.polyfit(list_x,list_y,degree)
        poly_y = np.poly1d(poly)(list_x)
        sns.lineplot(x=list_x, y=poly_y, label=index_country, 
                     sort=False, marker='o' if i_mark>=10 else '')
        i_mark = i_mark+1

# set country as index and plot new_data
def prepare_new_data(fp):
    new_data = pd.read_csv(fp)
    new_data = new_data.set_index("country")
    new_data = new_data.astype(int)
    return new_data
```

```python
fig = plt.figure(figsize=(14,16))
plt.yscale('log')
plt.xlabel("Days after the 100-th confirmed")
plt.ylabel("Logarithmic number of confirmed")

new_data = prepare_new_data(filepath_new_data_csv)
plt.xticks(np.arange(0, len(new_data.columns), step=5))
plot(new_data)
```

```python
# get population for every country in new_data
filepath_population = "./API_SP.POP.TOTL_DS2_en_csv_v2_887275/API_SP.POP.TOTL_DS2_en_csv_v2_887275.csv"

def prepare_population(fp):
    pop = pd.read_csv(fp)
    pop = pop[['Country Name', '2018']]
    pop = pop.set_index('Country Name')

    pop = pop.rename(columns={'2018': 'population'})
    pop = pop.rename(index={'United States': 'US'})
    pop = pop.rename(index={'Russian Federation': 'Russia'})
    pop = pop.rename(index={'Iran, Islamic Rep.': 'Iran'})
    pop = pop.rename(index={'Korea, Rep.': 'Korea, South'})
    # data_pop = pd.merge(new_data, pop, how='left', left_index=True, right_index=True).dropna()
    return pop
    
pop = prepare_population(filepath_population)

# compute percentage of confirmed for each country
new_data = prepare_new_data(filepath_new_data_csv)

def compute_rate(data, pop):
    for index_to_divide, row_to_divide in data.iterrows():
        try:
            data.loc[index_to_divide] = row_to_divide.map(lambda x: (100*x)/pop.loc[index_to_divide, 'population'])
        except:
            print("Impossible to get the population number for {}".format(index_to_divide))
    return data

new_data_rate = compute_rate(new_data, pop)
```

```python
plt.figure(figsize=(14,16))
plt.yscale('log')
plt.xlabel("Days after the 100-th confirmed")
plt.ylabel("Logarithmic rate of confirmed")
plt.xticks(np.arange(0, len(new_data_rate.columns), step=5))
plot(new_data_rate)
```

```python
DEGREE = 10
new_data = prepare_new_data(filepath_new_data_csv)

for i, r in new_data.iterrows():
    new_row = []
    for j in range(len(r)-1):
        new_row.append(r[j+1] - r[j] if r[j+1]>0 else -1)
    for _ in range(len(r)-len(new_row)):
        new_row.append(-1)
    new_data.loc[i] = new_row
    
plt.figure(figsize=(14,16))
plt.yscale('log')
plt.xlabel("Days after the 100-th confirmed")
plt.ylabel("lLogarithmic daily number of new confirmed")
plt.xticks(np.arange(0, len(new_data_rate.columns), step=5))
plot_smooth(new_data, DEGREE)
```
