```python
import pandas as pd
filepath = "./tests.csv"
tests = pd.read_csv(filepath)
tests = tests.set_index('Country')
tests
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit

#sns.scatterplot(x=tests['Total Tests'], y=tests['Confirmed'], hue=tests.index)
x = tests['Total Tests'][1:]
y = tests['Confirmed'][1:]
plt.figure(figsize=(20, 10))
plt.xlabel("Total number of tests")
plt.ylabel("Total number of confirmed")
plt.scatter(x, y)
for i, txt in enumerate(tests.index[1:]):
    plt.annotate(txt, (list(x)[i], list(y)[i]))
    
b, m = polyfit(x, y, 1)
plt.plot(x, b + m * x, '-')
```

```python
import numpy as np

plt.figure(figsize=(20, 10))
x = tests['Tests/1M pop']
plt.xlabel("Tests per milion of inhabitants")
y = (np.array(tests['Deaths'])/np.array(tests['Confirmed'])).tolist()
plt.ylabel("Mortality (rateo between deaths and confirmed)")

plt.scatter(x, y)
for i, txt in enumerate(tests.index):
    plt.annotate(txt, (list(x)[i], list(y)[i]))
    
b, m = polyfit(x, y, 1)
plt.plot(x, b + m * x, '-')
```
